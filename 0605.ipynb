## 데이터프레임 합치기

### 연결

```python
# 위아래로 합치기 concat(axis = 0)
# 좌우로 합치기 concat(axis = 1)
import pandas as pd
df1 = pd.DataFrame({'a': [1,2,3,4],
                    'b': [11,12,13,14],
                    'c': [21,22,23,24]})
df2 = pd.DataFrame({'a': [9,8,7],
                   'b': [19,18,17],
                   'c': [29,28,27],
                   'd': [39,38,37]})
print(df1)
print(df2)
```

```python
# 기본값 => axis = 0, join = 'outer'
# 위아래로 합치기 concat(axis = 0)
# 모든 열
res1 = pd.concat([df1,df2])
res1
```

```python
# axis = 0, join = 'inner'
# 공통된 열
res2 = pd.concat([df1,df2], join = 'inner')
res2
```

```python
# axis = 0, join = 'inner'
# 공통된 열 & 행 인덱스 초기화
res3 = pd.concat([df1,df2], join = 'inner', ignore_index=True)
res3
```

```python

# 좌우로 합치기 concat(axis = 1)
# 기본값 => join = 'outer'
# 모든 행
res4 = pd.concat([df1,df2], axis = 1)
res4
```

```python
# 공통된 행
res5 = pd.concat([df1,df2], axis = 1, join = 'inner')
res5
```

```python
# 데이터프레임과 시리즈 연결
sr1 = pd.Series([4,5,6,7], name = 'e')
sr2 = pd.Series([14,15,16,17], name = 'f', index = [1,2,3,4])
sr3 = pd.Series([24,25,26,27], name = 'g')
```

```python
print(sr1);print(sr2);print(sr3)
```

```python
# 시리즈를 axis = 1로 합치면 새로운 열을 합치는 것과 동일
# 시리즈의 이름이 열 이름
print(df1)
pd.concat([df1,sr1],axis = 1)
```

```python
# join = 'outer'
# 행 인덱스가 같은 것끼리 연결
# 모든 행
# 없는 인덱스인 경우에는 NaN
print(df2)
pd.concat([df2,sr2],axis = 1)
```

```python
# 행 인덱스가 같은 것끼리 연결
# 공통된 행만
print(df2)
pd.concat([df2,sr2], axis = 1, join = 'inner')
```

```python
# 인덱스 번호를 랜덤하게
df3 = pd.DataFrame({'a': [9,8,7],
                   'b': [19,18,17],
                   'c': [29,28,27],
                   'd': [39,38,37]}, index = [4,3,2])
```

```python
print(df3);print(sr2)
# df3 인덱스 번호 순서대로 합쳐짐 = 4, 3, 2, 1
pd.concat([df3,sr2], axis = 1)
```

```python
print(df3);print(sr2)
# sort = True 설정하면 인덱스 번호가 오름차순으로 정렬
pd.concat([df3,sr2], axis = 1, sort = True)
```

```python
# 시리즈와 시리즈를 연결
# axis = 0 위아래로 합치기
pd.concat([sr1,sr2], axis = 0)
```

```python
# join = 'outer'
# 모든 행
pd.concat([sr1,sr2], axis = 1)
```

### 병합

```python
# merge = SQL JOIN 과 유사 = 좌우로 합치기(병합)
# 공통된 열, 인덱스 => 키

# 라이브러리 불러오기
import pandas as pd
```

```python
df1 = pd.read_excel('/content/drive/MyDrive/HANARO/pandas datas/ch6/stock price.xlsx', engine = 'openpyxl')
df2 = pd.read_excel('/content/drive/MyDrive/HANARO/pandas datas/ch6/stock valuation.xlsx', engine = 'openpyxl')
print(df1.head())
print(df2.head())
```

```python
# INNER JOIN
# 기본값 how = 'inner', on = None (공통열)
merge_inner = pd.merge(df1,df2)
print(merge_inner.head())
```

```python
# FULL OUTER JOIN
merge_outer = pd.merge(df1,df2, how = 'outer', on = 'id')
print(merge_outer.head())
```

```python
# LEFT OUTER JOIN
merge_left = pd.merge(df1,df2, how = 'left', on = 'id')
print(merge_left.head())
```

```python

# RIGHT OUTER JOIN
merge_right = pd.merge(df1,df2, how = 'right', on = 'id')
print(merge_right.head())
```

```python
# 불 인덱스 => price < 100000
df1_price = df1[df1['price'] < 100000]
```

```python
# inner join
data = pd.merge(df1_price, df2)
print(data.head())
```

### 그룹 연산 메소드

#### 집계

```python
# 클래스 그룹 객체 생성
grouped = df.groupby('class')
```

```python
# 표준편차 <= 내장함수
gr_sd = grouped.std()
print(gr_sd)
print(type(gr_sd)) # 데이터프레임 = 여러 열
```

```python

gr_age_sd = grouped.age.std()
print(gr_age_sd)
print(type(gr_age_sd)) # 시리즈 = 하나의 열
```

```python
# 사용자 정의 함수
# 범위를 반환하는 함수
def range_xy(x):
  return x.max() - x.min()
```

```python
# 그룹 객체에 사용자 정의 함수 적용
grouped.agg(range_xy)
```

```python
# 함수를 여러 개 적용
grouped.agg(['min', 'max'])
```

```python
# 함수를 여러 개 적용 & 내가 원하는 열에 사용 가능
grouped.agg({'age': ['min', 'max'], # 나이의 최소값, 최대값
             'fare': 'mean'}) # fare 의 평균
```

#### 변환

```python
# transform = 그룹 연산 데이터 변환
# 단일 행 함수 = 데이터 개수와 함수 적용 결과 개수가 동일
age_mean = grouped.age.mean()
age_sd = grouped.age.std()
print(age_mean);print(age_sd)
```

```python
for i, j in grouped.age:
  z_score = (j - age_mean[i]) / age_sd[i]
  print("그룹 이름: ", i)
  print(z_score)
```

```python
def zscore(x):
  return (x - x.mean()) / x.std()
grouped.age.transform(zscore)
# 전체 데이터 개수 891 개 = First 개수 + Second 개수 + Third 개수
```

```python
# 계산이 가능한 모든 열에 대해서도 적용 가능
grouped.transform(zscore)
```

## 피봇 테이블

```python
# 1개의 행, 1개의 열, 1개의 값, 1개의 함수
pd.pivot_table(df, # 데이터
               index = 'class', # 행
               columns = 'sex', # 열
               values = 'age', # 값
               aggfunc = 'mean') # 집계 함수
```

```python
# 1개의 행, 1개의 열, 1개의 값, 2개의 함수
pd.pivot_table(df, # 데이터
               index = 'class', # 행
               columns = 'sex', # 열
               values = 'age', # 값
               aggfunc = ['mean', 'std']) # 집계 함수
```

```python
# 2개의 행, 1개의 열, 1개의 값, 2개의 함수
pd.pivot_table(df, # 데이터
               index = ['class', 'sex'], # 행
               columns = 'survived', # 열
               values = 'age', # 값
               aggfunc = ['mean', 'std']) # 집계 함수
```

```python
# 2개의 행, 1개의 열, 2개의 값, 2개의 함수
pv_table = pd.pivot_table(df, # 데이터
                          index = ['class', 'sex'], # 행
                          columns = 'survived', # 열
                          values = ['age', 'fare'], # 값
                          aggfunc = ['mean', 'std']) # 집계 함수
pv_table
```

```python
# 평균만 보기
pv_table.xs('mean', axis = 1)
# 데이터프레임
```
